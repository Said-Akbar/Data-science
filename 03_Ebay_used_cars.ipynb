## Ebay dataset of used cars

In this project, we are going to analyze a dataset of used cars from Ebay, which was scraped and cleaned to use by a Kaggle user. The link for the dataset is [here](https://www.kaggle.com/orgesleka/used-cars-database/data). The dataset contains around 50,000 records and 20 columns that includes car price, condition, year and more. Because the dataset is not clean, we will first do data cleaning and then dive into analyses.

```python
new_columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'n_pictures', 'postal_code',
       'last_seen']
```

```python
autos.head()
```

```python
autos.describe(include='all')
```

Above summary illustrates that 'seller' and 'offer_type' columns contain only one type of value for all records.
Anomalies:
'Price' contains some zero and NaN values. 'registration_year' also has a min value of 1000. There are Power_ps with zero values. ad_created contains value of 1946. 

```python
autos['price'] = autos['price'].str.replace('$','').str.replace(',','').astype(float)
```

```python
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',','').astype(float)
```

```python
autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)
```

We have converted 'price' and 'odometer' columns into a numeric datatype. This will facilitate our further analysis on outliers.

```python
print(autos['price'].unique().shape)
print('\nhighest prices:\n\n',autos['price'].value_counts().sort_index(ascending=False).head(10))
print('\nlowest prices:\n\n',autos['price'].value_counts().sort_index().head(10))

autos['price'].describe()
```

Based on the outliers information we can discard the outliers since they will distort our analysis. Furthermore, some of the prices are unreasonable high or low. By keeping a certain threshold we make sure our analysis is reasonable.

```python
autos = autos[autos['price'].between(500, 100000)]
```

```python
print(autos['odometer_km'].unique().shape)
print('\nhighest km:\n\n',autos['odometer_km'].value_counts().sort_index(ascending=False).head(10))
print('\nlowest km:\n\n',autos['odometer_km'].value_counts().sort_index().head(10))

autos['odometer_km'].describe()
```

Mileage data looks like within reasonable range: there are no huge or negative values.

Next, let us analyze the 'brand column to see the top, average and lowest selling car brands.

```python
autos['brand'].value_counts(sorted)
```

As we can see volkswagen, bmw and mecredes_benz are the most popular brands while trabant, lancia and lada are the least popular in German Ebay.

We will analyze the mean price of top 10 selling brands to see if they have substantial price differences.

```python
top10 = list(autos['brand'].value_counts(sorted).index[:10])
```

```python
mean_prices = {}
for brand in top10:
    one_brand = autos[autos['brand']==brand]
    mean_pr = one_brand['price'].mean()
    mean_prices[brand] = mean_pr
```

```python
mean_prices
```

Based on the dataset analysis, the highest mean price for cars are of audi brand and then mercedes_benz brand with 9571 and 8666 accordingly. The lowest mean prices belong to renault and then fiat with 2819 and 3256 prices accordingly. 

Mean mileage for top 10 brands

```python
mean_mileage = {}
for brand in top10:
    one_brand = autos[autos['brand']==brand]
    mean_km = one_brand['odometer_km'].mean()
    mean_mileage[brand] = mean_km
mean_mileage
```

```python
mean_prices = pd.Series(mean_prices)
mean_mileage = pd.Series(mean_mileage)
price_km = pd.DataFrame(mean_prices, columns=["mean_price"])
price_km["mean_mileage"] = mean_mileage

print(price_km)
```

Most of the cars from top 10 brands have over 110k km mileage and there are no substantial difference in their mileage on average. This means their prices are not influence by mileage.
